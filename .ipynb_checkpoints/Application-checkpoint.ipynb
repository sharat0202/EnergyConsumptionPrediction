```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Load the new dataset
new_data = pd.read_csv('EAE_Data_Test.csv', parse_dates=['timestamp'], index_col='timestamp')

# Fill missing values
new_data.fillna(method='ffill', inplace=True)

# Normalize the new data
scaler = MinMaxScaler()
scaled_new_data = scaler.fit_transform(new_data)

# Convert to DataFrame
scaled_new_df = pd.DataFrame(scaled_new_data, columns=new_data.columns, index=new_data.index)
scaled_new_df=scaled_new_df.drop(columns=['cpu-energy','memory-energy'])

def create_lag_features(df, lag=1):
    for col in df.columns:
        for i in range(1, lag + 1):
            df[f'{col}_lag_{i}'] = df[col].shift(i)
    df.dropna(inplace=True)
    return df

#---------------------------------------------------------
# Determine the number of rows to keep (top half)
num_rows = len(scaled_new_df)
num_rows_to_keep = num_rows // 2

# Select the top half of the DataFrame
df_top_half = scaled_new_df.iloc[:num_rows_to_keep]
# Select the bottom half of the DataFrame
df_bottom_half = scaled_new_df.iloc[num_rows_to_keep:]
#---------------------------------------------------------

# Create lag features for the new data
lagged_new_df = create_lag_features(df_top_half, lag=3)

# Define the features
X_new = lagged_new_df.drop(columns=['cpu', 'memory', 'sci-e'])

# Reshape data for LSTM [samples, timesteps, features]
X_new = X_new.values.reshape((X_new.shape[0], 1, X_new.shape[1]))
```

```python
# Load the pretrained model
model = tf.keras.models.load_model('time_series_model_2.h5')
# Predict
y_new_pred = model.predict(X_new)
```

```python
# Get the actual values
y_actual = lagged_new_df[['cpu', 'memory', 'sci-e']].values

# Evaluate
rmse_cpu = np.sqrt(mean_squared_error(y_actual[:, 0], y_new_pred[:, 0]))
rmse_memory = np.sqrt(mean_squared_error(y_actual[:, 1], y_new_pred[:, 1]))
rmse_sci_e = np.sqrt(mean_squared_error(y_actual[:, 2], y_new_pred[:, 2]))

print(f'RMSE (CPU): {rmse_cpu}')
print(f'RMSE (Memory): {rmse_memory}')
print(f'RMSE (SCI-E): {rmse_sci_e}')

# Plot predictions
plt.figure(figsize=(14, 8))
plt.plot(y_actual[:, 0], label='Actual CPU')
plt.plot(y_new_pred[:, 0], label='Predicted CPU')
plt.legend()
plt.show()

plt.figure(figsize=(14, 8))
plt.plot(y_actual[:, 1], label='Actual Memory')
plt.plot(y_new_pred[:, 1], label='Predicted Memory')
plt.legend()
plt.show()

plt.figure(figsize=(14, 8))
plt.plot(y_actual[:, 2], label='Actual SCI-E')
plt.plot(y_new_pred[:, 2], label='Predicted SCI-E')
plt.legend()
plt.show()
```
